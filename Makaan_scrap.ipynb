# Scrapping data from Makaan.com

```python
#importing required packages

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from IPython.display import clear_output
```

```python
owner_info=[]
no_BHK=[]
constr=[]
Ar=[]
loc=[]
Cost=[]
pr_per_sqft = []

base_url='https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city?_=1624536394722'

for i in range(1,51):
    
    time.sleep(3)
    print(f'Scrapping for page {i}')
    clear_output(wait=True)
    
    
    url=base_url+'&page='+str(i)
    req=requests.get(url)
    soup=BeautifulSoup(req.text,'html')
    
    #bhk info
    bhk_inf=soup.findAll('div',attrs={'class':'title-line'})
    for j in bhk_inf:
        no_BHK.append(j.strong.span.text)
    
    #location info
    loc_inf=soup.findAll('span',attrs={'itemprop':'addressLocality'})
    for k in loc_inf:
        loc.append(k.text)
    
    #owner_info
    own_inf=soup.findAll('span',attrs={'class':'seller-type'})
    for l in own_inf:
        owner_info.append(l.text)
    
    #price
    price=soup.findAll('div',attrs={'data-type':'price-link'})
    for m in price:
        Cost.append(m.text)
    
    #Area in sqft
    area=soup.findAll('td',attrs={'class':'size'})
    for n in area:
        Ar.append(n.span.text.strip())
        
    #construction status
    con_inf=soup.findAll('td',attrs={'class':'val'})
    for o in con_inf:
        constr.append(o.text)
        
    #price per sqft
    pps_info = soup.findAll('td',attrs={'class':'lbl rate'})
    for p in pps_info:
        pr_per_sqft.append(p.text)
    
```

```python
#creating Data Frame

datas=pd.DataFrame({'Construction':constr,
                  'Area_in_Sqft':Ar,
                  'Price':Cost,
                  'Owner_info':owner_info,
                  'Locality':loc,
                  'BHK':no_BHK,
                  'Price_per_sqft':pr_per_sqft})
datas
```

```python
# saving file with .csv extension
datas.to_csv('Makaan_scrapped_data.csv',index=False)
```
